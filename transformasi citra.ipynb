```python
import cv2
import numpy as np
# Import fungsi khusus dari Colab untuk menampilkan gambar
from google.colab.patches import cv2_imshow

# 1. Pastikan Anda sudah mengunggah gambar 'rio.jpeg' ke panel Files di sebelah kiri Colab
img = cv2.imread('balsa.jpg')

if img is not None:
    # Resize gambar
    t, l = img.shape[:2]
    img_resize = cv2.resize(img, (400, int((400 / l) * t)))

    # Konversi ke float untuk perhitungan logaritma pada setiap saluran warna
    img_float = img_resize.astype(np.float32)

    # Transformasi Logaritmik c = 36
    c1 = 36
    log_36 = np.clip(c1 * np.log1p(img_float), 0, 255).astype(np.uint8)

    # Transformasi Logaritmik c = 50
    c2 = 50
    log_50 = np.clip(c2 * np.log1p(img_float), 0, 255).astype(np.uint8)

    # Gabungkan gambar (Asli, Log 36, Log 50)
    hasil_gabungan = np.hstack((img_resize, log_36, log_50))

    # Tampilkan gambar di output cell Colab
    print("Urutan: Citra Asli | Log C=36 | Log C=50")
    cv2_imshow(hasil_gabungan)

else:
    print("Error: Gambar 'rio.jpeg' tidak ditemukan.")
    print("Silakan klik ikon folder di sebelah kiri layar Colab dan unggah file gambar Anda.")
```

```python
import cv2
import numpy as np
# Import library khusus Colab untuk menampilkan gambar
from google.colab.patches import cv2_imshow

# Membaca gambar tanpa atribut GRAYSCALE agar tetap berwarna (BGR)
# Pastikan Anda sudah mengunggah 'Image.jpg' ke Colab
img = cv2.imread('balsa.jpg')

if img is None:
    print("Gambar tidak ditemukan. Pastikan file 'Image.jpg' sudah diunggah ke panel Files.")
else:
    # Resize gambar agar muat di layar saat dijejer 3
    # Lebar diset ke 400 pixel, tinggi menyesuaikan rasio asli
    tinggi_asli, lebar_asli = img.shape[:2]
    lebar_baru = 400
    tinggi_baru = int((lebar_baru / lebar_asli) * tinggi_asli)
    img_resize = cv2.resize(img, (lebar_baru, tinggi_baru))

    # Transformasi Negative L = 255 pada citra warna
    L1 = 255
    neg_255 = L1 - img_resize

    # Transformasi Negative L = 128 pada citra warna
    L2 = 128
    # Mencegah underflow dengan int16, lalu membatasi nilai dan mengembalikan ke uint8
    neg_128 = np.clip(L2 - img_resize.astype(np.int16), 0, 255).astype(np.uint8)

    # Gabungkan gambar secara horizontal
    hasil_gabungan = np.hstack((img_resize, neg_255, neg_128))

    # Menampilkan hasil di Colab
    print("Urutan: Citra Asli | Negative 255 | Negative 128")
    cv2_imshow(hasil_gabungan)
```

```python
import cv2
import numpy as np
# Import library khusus Colab untuk menampilkan gambar
from google.colab.patches import cv2_imshow

# Pastikan Anda sudah mengunggah file 'Image.jpg' ke panel Colab
img = cv2.imread('balsa.jpg', cv2.IMREAD_GRAYSCALE)

if img is not None:
    # Ubah ke float32 untuk presisi perhitungan pangkat
    img_float = img.astype(np.float32)

    # Ambil 2 variasi saja agar muat di layar saat dijejer
    # Variasi 1: Gamma 0.9 (sedikit lebih terang)
    power_1 = np.clip(1.0 * np.power(img_float, 0.9), 0, 255).astype(np.uint8)

    # Variasi 2: Gamma 0.7 dan c=1.5 (jauh lebih terang)
    power_2 = np.clip(1.5 * np.power(img_float, 0.7), 0, 255).astype(np.uint8)

    # Gabungkan gambar secara horizontal
    hasil_gabungan = np.hstack((img, power_1, power_2))

    # Tampilkan di Colab
    print("Urutan: Citra Asli | Power Law (c=1, y=0.9) | Power Law (c=1.5, y=0.7)")
    cv2_imshow(hasil_gabungan)

else:
    print("Gambar tidak ditemukan. Pastikan file 'rio.jpeg' sudah diunggah.")
```

```python
import cv2
import numpy as np
# Import library khusus Colab untuk menggantikan cv2.imshow
from google.colab.patches import cv2_imshow

# Pastikan file 'rio.jpeg' sudah diunggah ke Colab
img = cv2.imread('balsa.jpg', cv2.IMREAD_GRAYSCALE)

if img is not None:
    img_float = img.astype(np.float32)
    # L adalah nilai intensitas maksimal untuk citra 8-bit
    L = 255

    # Inverse Logaritmik dengan c = 30
    c1 = 30
    inv_log_30 = np.clip(c1 * np.log1p(L - img_float), 0, 255).astype(np.uint8)

    # Inverse Logaritmik dengan c = 50
    c2 = 50
    inv_log_50 = np.clip(c2 * np.log1p(L - img_float), 0, 255).astype(np.uint8)

    # Menggabungkan gambar secara horizontal untuk perbandingan
    hasil_gabungan = np.hstack((img, inv_log_30, inv_log_50))

    # Menampilkan hasil di output cell Colab
    print("Urutan: Citra Asli | Inv Log (c=30) | Inv Log (c=50)")
    cv2_imshow(hasil_gabungan)

else:
    print("Gambar 'rio.jpeg' tidak ditemukan. Jangan lupa unggah file-nya dulu, ya!")
```
